```python
import numpy as np
import matplotlib.pyplot as plt

def f(x, p=0.5):
    """
    Función creciente y cóncava f(x) = x^p con 0 < p < 1.
    """
    return x**p

def simulate_st_petersburg(n_simulations=100000, p=0.5):
    """
    Simula el juego de San Petersburgo y calcula el valor esperado bajo f(x).
    """
    payouts = []
    for _ in range(n_simulations):
        n = 1
        while np.random.rand() > 0.5:  # Lanzamientos hasta que salga cara
            n += 1
        payouts.append(2**n)  # Pago 2^n
    
    # Aplicamos la función f(x)
    transformed_payouts = [f(x, p) for x in payouts]
    
    # Calculamos el valor esperado empírico
    expected_value = np.mean(transformed_payouts)
    return expected_value

# Simulamos y mostramos el resultado
p_values = np.linspace(0.1, 1, 10)
expected_values = [simulate_st_petersburg(p=p) for p in p_values]

plt.plot(p_values, expected_values, marker='o')
plt.xlabel('Exponente p en f(x) = x^p')
plt.ylabel('Valor esperado de f(X)')
plt.title('Valor esperado en la paradoja de San Petersburgo')
plt.grid()
plt.show()
```

```python
import numpy as np

def simulate_monty_hall(n_simulations=10000, switch=True):
    wins = 0
    for _ in range(n_simulations):
        prize_box = np.random.randint(0, 3)
        player_choice = np.random.randint(0, 3)

        remaining_boxes = [i for i in range(3) if i != player_choice and i != prize_box]
        revealed_box = np.random.choice(remaining_boxes)

        if switch:
            new_choice = [i for i in range(3) if i != player_choice and i != revealed_box][0]
        else:
            new_choice = player_choice

        if new_choice == prize_box:
            wins += 1

    return wins / n_simulations

stay_win_rate = simulate_monty_hall(switch=False)
switch_win_rate = simulate_monty_hall(switch=True)

print(f'Probabilidad de ganar manteniendo la caja: {stay_win_rate:.4f}')
print(f'Probabilidad de ganar cambiando de caja: {switch_win_rate:.4f}')
```
